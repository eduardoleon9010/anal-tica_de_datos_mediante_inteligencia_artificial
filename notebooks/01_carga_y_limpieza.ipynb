```python
# Importación de librerías esenciales para análisis de datos

import pandas as pd      # Manejo y análisis de datos en tablas
import numpy as np       # Operaciones numéricas eficientes
import plotly.express as px  # Visualizaciones interactivas
import plotly.io as pio      # Configuración global de estilo

# Establecer tema oscuro profesional para visualizaciones
pio.templates.default = "plotly_dark"
```

## Carga del Dataset


```python
# Cargar dataset desde GitHub (ubicación en tu repositorio)

url = "https://raw.githubusercontent.com/eduardoleon9010/anal-tica_de_datos_mediante_inteligencia_artificial/main/dataset/breast_cancer.csv"

df = pd.read_csv(url)

# Vista general de los primeros registros
df.head()
```

## Exploración

```python
# Información general del dataset
df.info()
```

```python
# Estadísticas descriptivas por columna numérica
df.describe()
```

## Limpieza de datos

```python
# Verificar si existen valores faltantes por columna
df.isnull().sum()
```

```python
# Renombrar variable objetivo para mejor interpretación
df.rename(columns={'target': 'diagnosis'}, inplace=True)

# Comprobación
df.head()
```

## Visualización

```python
# Histograma interactivo: Relación entre diagnóstico y radio medio

fig = px.histogram(
    df,
    x="mean radius",
    color="diagnosis",
    title="Distribución del Radio Medio según Diagnóstico"
)

fig.show()
```

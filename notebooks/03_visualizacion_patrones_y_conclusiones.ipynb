##  Objetivo

Implementar un modelo **Regresión Logística** para predecir el diagnóstico (benigno vs maligno) con base en variables clínicas del dataset.


### Cargar librerías

```python
# Importamos librerías necesarias para el modelado y análisis

import pandas as pd   # Manejo de datos tabulares
import numpy as np    # Cálculos matemáticos

# Herramientas de Machine Learning
from sklearn.model_selection import train_test_split   # Separar datos en entrenamiento y prueba
from sklearn.preprocessing import StandardScaler       # Estandarizar datos para mejorar el modelo
from sklearn.linear_model import LogisticRegression    # Algoritmo: Regresión Logística

# Evaluación del modelo
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

### Cargar Dataset

```python
# Cargar el dataset desde GitHub

url = "https://raw.githubusercontent.com/eduardoleon9010/anal-tica_de_datos_mediante_inteligencia_artificial/main/dataset/breast_cancer.csv"
df = pd.read_csv(url)

#  Visualizamos los primeros registros
df.head()
```

### Variables predictoras y objetivo

```python
#  X = características del tumor
#  y = diagnóstico (0 maligno, 1 benigno)

X = df.drop(columns=['diagnosis'])
y = df['diagnosis']

X.shape, y.shape
```

### División del dataset

```python
# Dividir datos en entrenamiento (75%) y prueba (25%)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

### Estandarización

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Entrenar Modelo

```python
# Estandarizar para que todas las variables usen misma escala
# Esto mejora el rendimiento de la Regresión Logística

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Entrenamiento

```python
model = LogisticRegression(max_iter=5000)
model.fit(X_train_scaled, y_train)
```

### Evaluación

```python
y_pred = model.predict(X_test_scaled)

print(" Exactitud del modelo:", accuracy_score(y_test, y_pred))
print("\n Reporte de Clasificación:\n", classification_report(y_test, y_pred))
```

### Matriz de confusión

```python
cm = confusion_matrix(y_test, y_pred)

import plotly.express as px
fig = px.imshow(
    cm,
    text_auto=True,
    color_continuous_scale="Blues",
    title="Matriz de Confusión - Regresión Logística",
    labels=dict(x="Predicción", y="Valor Real")
)
fig.show()
```

### Interpretación de coeficientes

```python
# Coeficientes del modelo (impacto de cada variable en el diagnóstico)

coef = pd.DataFrame({
    "feature": X.columns,
    "coeficiente": model.coef_[0]
})

coef_sorted = coef.sort_values(by="coeficiente", ascending=False)
coef_sorted
```

Interpretación:

Valores positivos → mayor probabilidad de ser benigno

Valores negativos → mayor probabilidad de ser maligno
___

Útil para decisiones clínicas

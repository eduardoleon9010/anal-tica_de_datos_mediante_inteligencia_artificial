### Cargar dataset

```python
#  Importar librerías necesarias

import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio

#  Tema oscuro profesional para visualizaciones
pio.templates.default = "plotly_dark"
```

```python
# Cargar el dataset directo desde GitHub

url = "https://raw.githubusercontent.com/eduardoleon9010/anal-tica_de_datos_mediante_inteligencia_artificial/main/dataset/breast_cancer.csv"
df = pd.read_csv(url)

df.head()
```

```python
# Renombrar la variable objetivo para estandarización del análisis
df.rename(columns={'target': 'diagnosis'}, inplace=True)
```

### Análisis del balance de clases

```python
# Cantidad de diagnósticos por categoría
conteo = df['diagnosis'].value_counts()
conteo
```

```python
# Visualización del balance de clases

fig = px.bar(
    conteo,
    title="Distribución del Diagnóstico (Benigno vs Maligno)",
    labels={'value': 'Cantidad', 'index': 'Diagnóstico'}
)
fig.show()
```

Interpretación:

Si hay desbalance, afecta modelos predictivos

Se deben considerar técnicas como sobremuestreo

### Distribución de características

```python
fig = px.box(
    df,
    x="diagnosis",
    y="mean radius",
    title="Diferencias del radio medio según diagnóstico"
)
fig.show()
```

### Correlaciones

```python
#  Matriz de correlación para variables numéricas
corr = df.corr()

#  Heatmap interactivo
fig = px.imshow(
    corr,
    title="Mapa de calor - Correlación entre variables"
)
fig.show()
```

## Conclusiones

Se identificaron patrones importantes:

| Hallazgo | Implicación analítica |
|---------|----------------------|
| Desbalance entre clases | Se requieren técnicas de ajuste en modelado |
| Variables relacionadas estructuralmente | Facilitan selección de características |
| Diferencias claras entre diagnósticos | Apoyan decisiones clínicas y preventivas |

Esto permite mejorar la precisión del diagnóstico en modelos predictivos,
cumpliendo con el objetivo del RA 220501115-02.

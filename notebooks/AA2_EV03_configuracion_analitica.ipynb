### Importar librerías

```python
# Importación de librerías para analítica de datos

import pandas as pd        # Manipulación de datos
import numpy as np         # Cálculos numéricos
import matplotlib.pyplot as plt  # Visualizaciones básicas
import seaborn as sns      # Gráficas estadísticas

print("Librerías cargadas correctamente")
```

### Cargar Dataset desde GitHub

```python
# Carga del dataset => Breast Cancer Wisconsin Diagnostic
url = "https://raw.githubusercontent.com/eduardoleon9010/anal-tica_de_datos_mediante_inteligencia_artificial/main/dataset/breast_cancer.csv"
df = pd.read_csv(url)

df.head()
```

### Exploración inicial

```python
print("Dimensiones del dataset:", df.shape)
df.info()
```

### Validación de calidad del dato

```python
print("Valores nulos por columna:")
df.isnull().sum()
```

### Estadísticas descriptivas

```python
df.describe()
```

### Visualización

```python
plt.figure(figsize=(8,5))
sns.histplot(df["mean radius"], kde=True)
plt.title("Distribución del radio del tumor")
plt.xlabel("Mean Radius")
plt.show()
```
